Q2. Write a python program to print Python Exception Hierarchy.

Answer - 2

```python
import inspect
from logging import *

LOG_FROMAT = "%(lineno)s * %(name)s * %(asctime)s * %(levelname)s * %(message)s"
basicConfig(filename="13_1_Feb.log", level=INFO, format=LOG_FROMAT)


def treeClass(cls, ind=0):

    str1 = "-"*ind
    str2 = cls.__name__
    info('%s %s', str1, str2)

    for i in cls.__subclasses__():
        treeClass(i, ind + 1)


info("Python Exception hierarchy is : ")

inspect.getclasstree(inspect.getmro(BaseException))

treeClass(BaseException)

shutdown()

with open("13_1_Feb.log") as f:
    print(f.read())
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Answer - 4

Lookup Error is the base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists.

KeyError

It arises when a key used to access a dictionaty item that is not present in the dictionary.

```python
d = {
    "name": "Aditya",
    "roll": 123
}

d[4]
```

IndexError

When we try to access an index that is out of range in the given iterable.

```python
l = ["Aditya", "Azad", "Data", "Science"]
l[5]
```
