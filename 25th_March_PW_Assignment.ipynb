### Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_excel("flight_price.xlsx")
```

```python
print(f"Rows = {df.shape[0]}, Columns = {df.shape[1]}")
```

### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'].tolist(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
print(f"Minimum Price : {df['Price'].min()}, Maximum Price : {df['Price'].max()}")
```

### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=df, palette='viridis')
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
df1.dropna(subset=['Rating', 'Category'], inplace=True)

# Create a boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Rating', data=df1, palette='viridis')
plt.title('App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.

```python
missing_values = df1.isnull().sum()

print("Missing values in the dataset:")
print(missing_values)
```

This code will output the number of missing values for each column in the dataset.

Missing values can impact the analysis in several ways:
1. **Data Integrity**: Missing values may compromise the integrity of the dataset, leading to incomplete or biased analysis results.
2. **Statistical Analysis**: Missing values can affect statistical measures such as mean, median, and standard deviation, potentially skewing the results.
3. **Visualization**: Missing values may affect the accuracy and interpretability of visualizations, leading to misleading insights.
4. **Model Performance**: Missing values can adversely affect the performance of machine learning models if not handled properly during preprocessing.

Depending on the extent and nature of missing values, strategies such as imputation (replacing missing values with estimated values) or removal of rows or columns with missing values may be necessary to mitigate their impact on the analysis. It's essential to carefully consider the implications of missing values and choose appropriate strategies to ensure the reliability and validity of the analysis results.

### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
df1 = pd.read_csv('Google_Playstore_Cleaned')
df1['Size'] = df1['Size'].astype(float)
df1.dropna(subset=['Rating', 'Size'], inplace=True)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df1['Size'], df1['Rating'], alpha=0.5, color='blue')
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (KB)')
plt.ylabel('Rating')
plt.show()
```

### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df1['Price'] = df1['Price'].astype(float)
df1.dropna(subset=['Type', 'Price'], inplace=True)

df1.groupby(by='Type').mean()['Price']

df1.groupby(by='Type').mean()['Price'].plot(kind='bar',title='Barchart of app Type vs Average Price')
```
