### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular Python visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot:
   - Use: Scatter plots are used to visualize the relationship between two continuous variables. They are useful for identifying patterns, clusters, or correlations in the data.

2. Line Plot:
   - Use: Line plots are suitable for displaying trends or changes over time. They are often used to visualize time-series data or any data where one variable changes with another.

3. Histogram:
   - Use: Histograms are used to represent the distribution of a single numerical variable. They help in understanding the underlying frequency distribution of the data, such as identifying peaks, skewness, or central tendencies.

4. Bar Plot:
   - Use: Bar plots are ideal for comparing categorical data. They display the relationship between a categorical variable and a numerical variable, showing the mean or count of the numerical variable for each category.

5. Heatmap:
   - Use: Heatmaps are effective for visualizing the matrix-like data. They use color gradients to represent the magnitude of values in a matrix, making it easier to identify patterns or correlations between variables.

These are just a few examples, and Seaborn offers many more plot types for various data visualization needs.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x='timepoint', y='signal', data=fmri,hue='event',style='region')
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
### Note: pclass, age, and fare are columns in the titanic dataset.

```python
titan=sns.load_dataset('titanic')
```

```python
sns.boxplot(x='pclass',y='age',data=titan)
```

```python
sns.boxplot(x='pclass',y='fare',data=titan)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
```

```python
import matplotlib.pyplot as plt
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger counts by month and year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
