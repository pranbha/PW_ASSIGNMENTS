### Q1. What is the Probability density function?

The probability density function (PDF) is a fundamental concept in probability theory and statistics. It describes the likelihood of a continuous random variable taking on a particular value within a given range. 

Mathematically, the PDF of a continuous random variable \( X \) is denoted as \( f(x) \) and is defined such that the probability of \( X \) lying between two values \( a \) and \( b \) is given by the integral of the PDF over that interval.

Properties of the PDF:

1. The PDF must be non-negative for all values of the random variable.
2. The total area under the PDF curve over the entire range of possible values must be equal to 1


The PDF is often used to calculate probabilities and expected values in continuous probability distributions, such as the normal distribution, exponential distribution, and many others.

### Q2. What are the types of Probability distribution?

There are several types of probability distributions, each with its own characteristics and applications. Here are some common ones:

1. Discrete Probability Distributions:
   - Bernoulli Distribution: Describes the probability of success or failure in a single experiment.
   - Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.
   - Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence.

2. Continuous Probability Distributions:
   - Normal Distribution (Gaussian Distribution): Symmetric bell-shaped distribution that is widely used in natural and social sciences due to the central limit theorem.
   - Uniform Distribution: All values within a given interval are equally likely.
   - Exponential Distribution: Describes the time between events in a Poisson process, such as the time between phone calls at a call center.
   - Gamma Distribution: Generalizes the exponential distribution to represent the sum of \( k \) exponential random variables.
   - Beta Distribution: Describes the distribution of probabilities in the interval [0, 1], often used in Bayesian statistics.
   - Chi-Square Distribution: Distribution of the sum of squared standard normal deviates, commonly used in hypothesis testing and confidence interval construction.
   - Student's t-Distribution: Used for estimating population parameters when the sample size is small and the population variance is unknown.
   - F-Distribution: Used in the analysis of variance (ANOVA) and in regression analysis.

These are some of the most commonly encountered probability distributions, each with its own properties and applications in various fields such as engineering, finance, biology, and more.

### Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Args:
    - x: The point at which to evaluate the PDF.
    - mean: The mean of the normal distribution.
    - std dev: The standard deviation of the normal distribution.

    Returns:
    The PDF value at the given point x.
    """
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    prefactor = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf = prefactor * math.exp(exponent)
    return pdf

# Example usage:
mean = 0
std_dev = 1
x = 1
pdf_value = normal_pdf(x, mean, std_dev)
print("PDF at x =", x, ":", pdf_value)
```

### Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. The properties of the binomial distribution include:

1. Fixed Number of Trials: The number of trials, denoted by \( n \), is fixed in advance.

2. Independent Trials: Each trial is independent of the others.

3. Two Possible Outcomes: Each trial results in one of two outcomes, often labeled as success (with probability \( p \)) or failure (with probability \( 1 - p \)).

4. Constant Probability: The probability of success, denoted by \( p \), remains constant from trial to trial.

5. Discrete Values: The binomial random variable \( X \) represents the number of successes in \( n \) trials, and it takes on discrete values ranging from 0 to \( n \).

Examples of events where the binomial distribution can be applied include:

1. Coin Flips: Suppose you flip a fair coin \( n \) times and count the number of heads. Each flip is a Bernoulli trial with a probability of success (heads) equal to \( p = 0.5 \), and the total number of heads follows a binomial distribution.

2. Drug Trials: In a clinical trial, patients are often classified as responders or non-responders based on whether a certain treatment is effective for them. If the probability of response to the treatment is known, the number of responders among a group of \( n \) patients can be modeled using a binomial distribution.

In both examples, the binomial distribution helps quantify the likelihood of observing a certain number of successes (e.g., heads, responders) out of a fixed number of trials (e.g., coin flips, patients in a trial).

### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generate random sample from binomial distribution
random_sample = np.random.binomial(n, p, size=n)

# Plot histogram
plt.hist(random_sample, bins=range(n+1), density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Sample from Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

### Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Args:
    - k: The point at which to evaluate the CDF.
    - mean: The mean (lambda) of the Poisson distribution.

    Returns:
    The CDF value at the given point k.
    """
    cdf = 0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean ** i) / math.factorial(i)
    return cdf

# Example usage:
mean = 3
k = 2
cdf_value = poisson_cdf(k, mean)
print("CDF at k =", k, ":", cdf_value)
```

### Q7. How Binomial distribution different from Poisson distribution?

Binomial and Poisson distributions are both used to model the number of successes in a series of independent trials, but they differ in several key aspects:

1. Number of Trials:
   - Binomial Distribution: The number of trials, denoted by \( n \), is fixed and known in advance.
   - Poisson Distribution: There is no fixed number of trials. Instead, the distribution models the number of events occurring in a fixed interval of time or space.

2. Nature of Trials:
   - Binomial Distribution: Each trial has two possible outcomes (success or failure).
   - Poisson Distribution: The outcome of each event is either success or failure, but the number of events can be large and potentially infinite.

3. Probability of Success:
   - Binomial Distribution: The probability of success in each trial, denoted by \( p \), remains constant from trial to trial.
   - Poisson Distribution: There is no direct notion of a probability of success in individual events. Instead, the distribution is characterized by the average rate of events occurring within the fixed interval, denoted by \( \lambda \).

4. Parameterization:
   - Binomial Distribution: Parameterized by two values: the number of trials \( n \) and the probability of success \( p \).
   - Poisson Distribution: Parameterized by a single value: the average rate \( \lambda \) at which events occur in the fixed interval.

### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# Parameters
sample_size = 1000
mean = 5

# Generate random sample from Poisson distribution
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

### Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In both the Binomial and Poisson distributions, the mean (\( \mu \)) and variance (\( \sigma^2 \)) are closely related, but their relationship differs slightly between the two distributions.

Binomial Distribution:
1. Mean (\( \mu \)): For a binomial distribution with parameters \( n \) (number of trials) and \( p \) (probability of success), the mean is given by \( \mu = np \).
2. Variance (\( \sigma^2 \)): The variance of a binomial distribution is given by \( \sigma^2 = np(1-p) \).

The relationship between the mean and variance in the binomial distribution is characterized by the product of \( np(1-p) \). As \( p \) approaches 0 or 1, the variance decreases, reaching its minimum when \( p = 0.5 \). The variance is symmetric around the mean.

Poisson Distribution:
1. Mean (\( \mu \)): For a Poisson distribution, the mean and the variance are both equal to \( \lambda \), the average rate of events occurring in a fixed interval of time or space.
2. Variance (\( \sigma^2 \)): Similarly, the variance of a Poisson distribution is also equal to \( \lambda \).

In the Poisson distribution, the mean and variance are both represented by \( \lambda \), indicating that the spread of the distribution is directly proportional to the mean. As \( \lambda \) increases, the distribution becomes more spread out.

### Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, also known as a Gaussian distribution, the least frequent data points occur at the tails of the distribution, farthest away from the mean. Specifically, the least frequent data points are located in the regions of the distribution that are several standard deviations away from the mean.

In a standard normal distribution (with a mean of 0 and a standard deviation of 1), approximately 68% of the data falls within one standard deviation of the mean, about 95% falls within two standard deviations, and around 99.7% falls within three standard deviations. Therefore, the least frequent data points, representing approximately 0.3% of the total, are found more than three standard deviations away from the mean, in the tails of the distribution.

In a normal distribution with a non-zero mean, the least frequent data points are similarly located in the tails of the distribution, but their exact positions depend on the mean and standard deviation of the distribution. The farther the data points are from the mean, the less frequent they are in a normal distribution.
